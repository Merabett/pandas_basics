#**GROUPING**

```python
titanic.groupby(['pclass'])[['survived','fare']].sum()   #a la izquierda las columnas por las que queremos clasificar, a la derecha las columnas que se estudian y
                                                        # a las cuales aplicamos funciones como max, min...
```

```python
titanic.groupby(['pclass','survived'])[['fare']].sum()
```

```python
titanic.groupby(['pclass','sex','survived'])[['fare']].agg(['max','sum'])
```

```python
groups=titanic.groupby(['pclass','sex','survived'])[['fare','ticket']].agg(['max','sum'])
groups
```

```python
#getting rid of multi-index with reset_index()
notmultiindex=groups.reset_index()
notmultiindex
```

#**SORTING**

```python
groups.sort_index()  #default
```

```python
group_2=titanic.groupby(['pclass','sex']).fare.agg(['sum'])
group_2  #sort_index
```

```python
group_2.sort_values(by='sum')
```
